```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup

```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python

table = soup.find_all('table')[0] 
df = pd.read_html(str(table))[0]
df.columns = df.iloc[0]
df = df.iloc[1:]
df
```

```python
df2 = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(lambda x: ','.join(x.astype(str))).reset_index()
df2[df2['Borough']=='Toronto']
```

```python
df3 = df2
```

```python
def getLat(par1):
    # initialize your variable to None
    lat_lng_coords = None

    # loop until you get the coordinates
    while(lat_lng_coords is None):
      g = geocoder.arcgis ('{}, Toronto, Ontario'.format(par1))
      lat_lng_coords = g.latlng

    return lat_lng_coords[0]
```

```python
def getLon(par1):
    # initialize your variable to None
    lat_lng_coords = None

    # loop until you get the coordinates
    while(lat_lng_coords is None):
      g = geocoder.arcgis ('{}, Toronto, Ontario'.format(par1))
      lat_lng_coords = g.latlng

    return lat_lng_coords[1]
```

```python
df3['Latitude'] = df3.apply(lambda x: getLat(x['Postcode']), axis=1)
df3['Longitude'] = df3.apply(lambda x: getLon(x['Postcode']), axis=1)
```
